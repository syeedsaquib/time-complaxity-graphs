```python
import numpy as np
import matplotlib.pyplot as plt
import math
a = [i for i in range(1,100)]
n = np.array(a)
```

#### O(N) time complaxity graph

```python
plt.plot(n)
print(a,n)
plt.show()
```

#### O(n^2) time complaxity graph

```python
n_square = n**2
print(n_square)
plt.plot(a,n_square)
plt.show()
```

#### O(!n) time complaxity graph 

```python
fact_n = [math.factorial(i) for i in range(1,100)]
print(fact_n)
plt.plot(a,fact_n)
plt.show()
```

#### O(1) time complaxity graph

```python
ones = [i**0 for i in range(1,100)]
print(ones)
plt.plot(a,ones)
plt.show()
```

#### O(log(n)) time complaxity graph

```python
log_n = [math.log2(i) for i in range(1,100)]
print(log_n)
plt.plot(a,log_n)
plt.show()
```

#### O(2^n) time complaxity graph

```python
two_pow_n = [math.pow(2,i) for i in range(1,100)]
print(two_pow_n)
plt.plot(a,two_pow_n)
plt.show()
```

#### O(n*log(n)) time complaxity graph

```python
two_log_n = [i*(math.log2(i)) for i in range(1,100)]
print(two_log_n)
plt.plot(a,two_log_n)
plt.show()
```
